```python
class Film:
    def __init__(self, ad, yonetmen, yapim_tarihi, tur):
        self.ad = ad
        self.yonetmen = yonetmen
        self.yapim_tarihi = yapim_tarihi
        self.tur = tur

    def __str__(self):
        return f"Film Adı = {self.ad}\nYönetmeni = {self.yonetmen}\nYapım Tarihi = {self.yapim_tarihi}\nTürü = {self.tur}"

class FilmKayit:
    def __init__(self, dosya_adi):
        self.dosya_adi = dosya_adi

    def film_ekle(self):
        film_ad = input("Film adını girin: ")
        yonetmen = input("Yönetmeni girin: ")
        yapim_tarihi = input("Yapım tarihini girin: ")
        tur = input("Filmin türünü girin: ")
        film = Film(film_ad, yonetmen, yapim_tarihi, tur)

        with open(self.dosya_adi, 'a') as dosya:
            dosya.write(str(film) + '\n')
        print(f"{film.ad} kaydedildi.")

    def film_ara(self):
        film_ad = input("Aramak istediğiniz film adını girin: ")
        with open(self.dosya_adi, 'r') as dosya:
            for line in dosya:
                if film_ad.lower() in line.lower():
                    print(line.strip())

    def calistir(self):
        while True:
            print("1- Film Ekle")
            print("2- Film Ara")
            print("q- Çıkış")
            secim = input("Seçiminizi yapın: ")

            if secim == '1':
                self.film_ekle()
            elif secim == '2':
                self.film_ara()
            elif secim == 'q':
                break
            else:
                print("Geçersiz seçim. Tekrar deneyin.")


film_kayit = FilmKayit("film_kayit.txt")
film_kayit.calistir()


with open("film_kayit.txt", 'r') as dosya:
    for line in dosya:
        print(line.strip())
        print()
```
